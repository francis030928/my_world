# ***LINEAR REGRESSION***

Linear Regression is a statistical method used to establish a relationship between a dependent variable and one or more independent variables by fitting a linear equation to observed data points. The goal of linear regression is to minimize the residual sum of squares between the observed response values (y) and the values predicted by the linear approximation (ŷ).<br>


Note that linear regression makes several assumptions about the data, such as linearity, independence, homoscedasticity, and normality. These assumptions should be carefully checked before and after fitting the model.<br> <br>
What is a slope?
In a regression context, the slope is very important in the equation because it tells you how much you can expect Y to change as X increases.
It is denoted by m in the formula y = mx+b.<br>
What is Intercept?
The y-intercept is the place where the regression line y = mx + b crosses the y-axis (where x = 0), and is denoted by b.<br><br>
How does a linear Regression Work? 
The whole idea of the linear Regression is to find the best fit line,which has very low error(cost function).
This line is also called Least Square Regression Line(LSRL).

The line of best fit is described with the help of the formula y=mx+b.
where,m is the Slope and
b is the intercept.

Properties of the Regression line:
1. The line minimizes the sum of squared difference between the observed values(actual y-value) and the predicted value(ŷ value)
2. The line passes through the mean of independent and dependent features.

<h1>Key assumptions of effective linear regression</h1>

Assumptions to be considered for success with linear-regression analysis:
<ul>
<li>For each variable: Consider the number of valid cases, mean and standard deviation. </li>
<li>For each model: Consider regression coefficients, correlation matrix, part and partial correlations, multiple R, R2, adjusted R2, change in R2, standard error of the estimate, analysis-of-variance table, predicted values and residuals. Also, consider 95-percent-confidence intervals for each regression coefficient, variance-covariance matrix, variance inflation factor, tolerance, Durbin-Watson test, distance measures (Mahalanobis, Cook and leverage values), DfBeta, DfFit, prediction intervals and case-wise diagnostic information. </li>
<li>Plots: Consider scatterplots, partial plots, histograms and normal probability plots.</li>
<li>Data: Dependent and independent variables should be quantitative. Categorical variables, such as religion, major field of study or region of residence, need to be recoded to binary (dummy) variables or other types of contrast variables.  </li>
<li>Other assumptions: For each value of the independent variable, the distribution of the dependent variable must be normal. The variance of the distribution of the dependent variable should be constant for all values of the independent variable. The relationship between the dependent variable and each independent variable should be linear and all observations should be independent.</li>
</ul>

Make sure your data meets linear-regression assumptions
Before you attempt to perform linear regression, you need to make sure that your data can be analyzed using this procedure. Your data must pass through certain required assumptions.

Here’s how you can check for these assumptions:
<ul>
<li>The variables should be measured at a continuous level. Examples of continuous variables are time, sales, weight and test scores. </li>
<li>Use a scatterplot to find out quickly if there is a linear relationship between those two variables.</li>
<li>The observations should be independent of each other (that is, there should be no dependency).</li>
<l1>Your data should have no significant outliers. </li>
<li>Check for homoscedasticity — a statistical concept in which the variances along the best-fit linear-regression line remain similar all through that line.</li>
<li>The residuals (errors) of the best-fit regression line follow normal distribution. </li> 
</ul>

In a linear regression model, the coefficient (also known as the weight or slope) and the intercept represent the parameters that define the linear relationship between the independent variable(s) and the dependent variable.

The equation of a linear regression line can be expressed as:

ŷ = b0 + b1 * X1 + b2 * X2 + ... + bn * Xn

Where:

ŷ is the predicted value of the dependent variable
X1, X2, ..., Xn are the independent variables
b0 is the intercept
b1, b2, ..., bn are the coefficients (slopes)
The intercept, b0, represents the value of ŷ when all the independent variables are zero. It gives the baseline value for the dependent variable and can be interpreted as the expected value of ŷ when all independent variables are 0.

The coefficients, b1, b2, ..., bn, represent the change in the dependent variable associated with a unit change in the corresponding independent variable, holding all other independent variables constant. The coefficients can be used to determine the relative importance of each independent variable in explaining the variation in the dependent variable.

The coefficient and intercept values are estimated during the model training process using a optimization algorithm such as gradient descent or least squares. The goal is to find the values that minimize the residual sum of squares between the observed and predicted values.

In summary, the coefficient and intercept are useful in linear regression because they allow us to quantify the relationship between the independent variables and the dependent variable, and make predictions for new data points based on the fitted model.

````python
# Check for Linear Relationship between features and label

def Linearity_test (dataframe, a, b, c, d, dependent):
    plt.figure(figsize =(30,4))
    sns.set(style = 'darkgrid')
    plt.subplot(1,5,1)
    plt.scatter(dataframe[a], dataframe[dependent])
    plt.title(a + ' ScatterPlot')
    plt.xlabel(a)
    plt.ylabel(dependent)
    plt.subplot(1,5,2)
    plt.scatter(dataframe[b], dataframe[dependent])
    plt.title(b + ' ScatterPlot')
    plt.xlabel(b)
    plt.subplot(1,5,3)
    plt.scatter(dataframe[c], dataframe[dependent])
    plt.title(c + ' ScatterPlot')
    plt.xlabel(c)
    plt.subplot(1,5,4)
    plt.scatter(dataframe[d], dataframe[dependent])
    plt.title(d + ' ScatterPlot')
    plt.xlabel(d)
    plt.show()

Linearity_test(df, 'R&D Spend', 'Administration', 'Marketing Spend', 'State', 'Profit')
````

We can assume there is a linear relationship between the all the features except the state feature which shows no linear relationship, so we drop it.

````python
# Drop the state column has it has no linear relationship with the Dependent Variable
df_scaled.drop(['State'], axis = 1, inplace = True)
df_scaled.columns
````

````python
# Check for outliers using the boxplot
def outlier_check (dataframe, a, b, c, dependent):
    plt.figure(figsize =(30,4))
    sns.set(style = 'darkgrid')
    plt.subplot(1,5,1)
    plt.boxplot(dataframe[a])
    plt.title(a + ' boxPlot')
    plt.xlabel(a)
    plt.ylabel(dependent)
    plt.subplot(1,5,2)
    plt.boxplot(dataframe[b])
    plt.title(b + ' boxPlot')
    plt.xlabel(b)
    plt.subplot(1,5,3)
    plt.boxplot(dataframe[c])
    plt.title(c + ' boxPlot')
    plt.xlabel(c)
    plt.show()

outlier_check(df, 'R&D Spend', 'Administration', 'Marketing Spend', 'Profit')
````

<H1>POLYNOMIAL REGRESSION</H1>

Polynomial Regression is a regression algorithm that models the relationship between a dependent(y) and independent variable(x) as nth degree polynomial. The Polynomial Regression equation is given below:


```
y = b0 + b1X1 + b2X1^2 + b2X1^3 + ... bnX1n + b2X2^1 + b2X2^2 + ...bnXn^n
```
Polynomial Regression does not require the relationship between the independent and dependent variables to be linear in the data set. This is also one of the main difference between the Linear and Polynomial Regression.
Polynomial Regression is generally used when the points in the data are not captured by the Linear Regression Model and the Linear Regression fails in describing the best result clearly.
As we increase the degree in the model,it tends to increase the performance of the model. However, increasing the degrees of the model also increases the risk of over-fitting and under-fitting the data.<br>
It is also important to note that all regression are polynomial in nature, only that their degree varies. A simple or multiple linear regression a polynomial regression of degree 1, while the polynomial regression of degree higher than 1 <br>

Information worth knowing about the Polynomial Regression:<ul>
<li>It is also called the special case of Multiple Linear Regression in ML. Because we add some polynomial terms to the Multiple Linear regression equation to convert it into Polynomial Regression.
<li>It is a linear model with some modification in order to increase the accuracy. The dataset used in Polynomial regression for training is of non-linear nature, and thus makes use of a linear regression model to fit the complicated and non-linear functions and datasets.</li>
<li>Overfitting: One common problem in polynomial regression is overfitting, which occurs when the model fits the training data too well and becomes too complex. This can lead to poor generalization performance on unseen data.</li>
<li>Regularization: To prevent overfitting, regularization techniques such as Ridge Regression or Lasso Regression can be used, which add a penalty term to the loss function being optimized to reduce the magnitude of the coefficients.</li>
<li>Model Selection: The choice of the degree of the polynomial is an important step in polynomial regression. A high degree polynomial can lead to overfitting, while a low degree polynomial may not capture the underlying relationship between the variables. Model selection can be done using techniques such as cross-validation.</li>
<li>Non-Linear Relationships: Polynomial regression can be used to model non-linear relationships between the independent and dependent variables. This makes it useful for a wide range of applications, including modeling trends in financial data, analyzing the effect of different factors on a response variable, etc.</li>
<li>Limitations: While polynomial regression can model complex non-linear relationships, it can also be sensitive to the presence of outliers or irrelevant variables in the data. Additionally, polynomial regression models can be computationally expensive, especially for high-degree polynomials.</li>
