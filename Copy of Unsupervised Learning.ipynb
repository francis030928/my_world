<h1>CLUSTERING</h1>
<hr>

Clustering is a type of unsupervised machine learning algorithm that is used to group similar data points together based on their shared characteristics. The goal of clustering is to divide a set of data points into distinct groups, or clusters, in such a way that the data points within a cluster are more similar to each other than to those in other clusters.<br>
Clustering algorithms work by analyzing the data and identifying patterns and similarities between the data points. There are many different types of clustering algorithms, but they all involve some method of measuring the similarity or dissimilarity between data points and then grouping the most similar points together.<br>
Clustering can be used in a variety of fields, including marketing, biology, social sciences, and computer science. For example, in marketing, clustering can be used to identify groups of customers with similar buying habits or demographics, while in biology, clustering can be used to group genes or proteins based on their functions or properties. Overall, clustering is a powerful tool for understanding and organizing large datasets, and it can be used to reveal insights and patterns that may not be immediately apparent from simply looking at the raw data.<br><br>
There are several types of clustering algorithms that can be used depending on the type of data being analyzed and the specific goals of the analysis. Here are some common types of clustering:
<ul><li>
K-Means Clustering: Its a <b>Partitioning Method</b> of clustering algorithm that involves partitioning the data into K clusters, where K is a pre-defined number. The algorithm works by iteratively updating the cluster centroids until the clusters are optimized. For example, K-means clustering can be used to segment customers based on their purchase behavior, demographic data, and other relevant factors. This helps businesses to identify distinct customer groups and develop tailored marketing strategies for each group. Another example is that K-means clustering can be used to segment products based on their characteristics, such as price, quality, and features. This helps businesses to identify different product categories and optimize their pricing and marketing strategies for each category.</li>
<li>
Agglomerative Clustering: Its an <b>Hierarchical</b> of clustering which involves grouping data points into a hierarchical tree-like structure, where similar data points are grouped together at each level. Hierarchical clustering can be either agglomerative (merging clusters together) or divisive (splitting clusters apart).</li>
<li>
Density-based Clustering: Density-based clustering is a type of clustering algorithm that groups data points based on their proximity to each other in dense regions of the data space. The most common density-based clustering algorithm is DBSCAN.</li>
<li>
Fuzzy Clustering: Fuzzy clustering is a clustering algorithm that allows data points to belong to multiple clusters, with each point assigned a probability of belonging to each cluster. Fuzzy clustering can be useful when data points have ambiguous or overlapping features.</li>
<li>
Model-based Clustering: Model-based clustering involves fitting a statistical model to the data and then clustering the data based on the parameters of the model. Examples of model-based clustering algorithms include Gaussian Mixture Models (GMM) and Hidden Markov Models (HMM).</li></ul><br>

We will be focusing on Agglomerative and K-Means Clustering for the purpose of this session.<br><br>
<b>Agglomerative Clustering:</b> Agglomerative clustering is a type of clustering algorithm used in unsupervised machine learning. Agglomerative Hierarchical Clustering uses a bottom-up approach to form clusters. That means it starts from single data points. Then it clusters the closer data points into one cluster. The same process repeats until it gets one single cluster.<br>
The algorithm starts by considering each data point as its own cluster, and then iteratively merges the two closest clusters together, continuing until all data points are in a single cluster or until a stopping criterion is met. This merging process is based on the distance between the clusters, which is typically calculated as the distance between their centroids or the minimum distance between any two points in the two clusters.<br>
Divisive Hierarchical Clustering is the opposite of Agglomerative Hierarchical clustering. It is a Top-Down approach. That means, it starts from one single cluster. In that single cluster, there may be n number of clusters and data points. At each step it split the farthest cluster into separate clusters.<br><br>
<b>K-means clustering:</b> It's a popular unsupervised machine learning algorithm that is used to partition a given set of data points into K clusters, where K is a pre-defined number. The algorithm works by iteratively optimizing the sum of squared distances between the data points and their nearest cluster centroids.<br>
The K-means clustering algorithm starts by randomly initializing K cluster centroids within the data space. It then proceeds to iterate between two steps:<br>
---- Assign each data point to its nearest cluster centroid. This step involves computing the distance between each data point and each cluster centroid, and assigning each data point to the cluster whose centroid is nearest to it.
<br>
---- Update each cluster centroid to the mean of the data points assigned to it. This step involves computing the mean of all the data points assigned to each cluster, and updating the cluster centroid to this mean value.
<br>
The algorithm continues to iterate between these two steps until the cluster assignments no longer change or a maximum number of iterations is reached. At the end of the algorithm, each data point will be assigned to one of the K clusters.
<br>
K-means clustering is widely used in various fields such as data mining, machine learning, image processing, and natural language processing. It can be applied to a wide range of datasets and is particularly useful when the number of clusters is known a priori, and the data points are well-separated.<br><br><br>

<h2><b>PREPROCESSING</b></h2>

```python
# Standard Scale the Numerical variables
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

x = frame[['Age', 'Income', 'Score']]
for i in x.columns:
  x[[i]] = scaler.fit_transform(x[[i]])

x.head()
```

<h1><b><u>K -MEANS CLUSTERING</u></b></h1>

```python
#install yellowbrick to vizualize the Elbow curve
!pip install yellowbrick  --q

from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer

x = frame[['Income', 'Score']]

# Instantiate the clustering model and visualizer
km = KMeans(random_state=42)
visualizer = KElbowVisualizer(km, k=(2,10))
 
plt.figure(figsize = (10, 5))
visualizer.fit(x) #............................................................. Fit the data to the visualizer
visualizer.show() #............................................................. Finalize and render the figure
```

```python
# Having gotten the optimal number of K, we fit kmeans on our dataset.
kmeans_5 = KMeans(n_clusters = 5).fit(x) # ..................................... Fit Kmeans Algorthm on the Data Subset

X = x.values
k_means_label = kmeans_5.labels_ # ............................................. Instantiate the labels of the model

# Plot the dataset using the model labelling to colour code them
plt.figure(figsize = (12, 6)) 
sns.set(style = 'darkgrid')
plt.scatter(X[k_means_label ==0,  0], X[k_means_label == 0, 1], s = 100, c = 'blue', label = 'Cluster 1')  
plt.scatter(X[k_means_label == 1, 0], X[k_means_label == 1, 1], s = 100, c = 'green', label = 'Cluster 2')  
plt.scatter(X[k_means_label== 2, 0], X[k_means_label == 2, 1], s = 100, c = 'red', label = 'Cluster 3')  
plt.scatter(X[k_means_label == 3, 0], X[k_means_label == 3, 1], s = 100, c = 'yellow', label = 'Cluster 4')  
plt.scatter(X[k_means_label == 4, 0], X[k_means_label == 4, 1], s = 100, c = 'magenta', label = 'Cluster 5')  
plt.title('Clusters of customers')  
plt.xlabel('Income')  
plt.ylabel('Spending Score (1-100)')  
plt.legend()  
plt.show() 
```

<u>5 proves to be an optimal number for clustering. So we save the model.</u><br><br>
<b>Cluster Interpretation</b>: <br>
- <u>Class Yellow</u>: This is the class of customers that earns low income, and also makes lesser purchases as a result of their low income. Promotional adverts should be targetted at them.
- <u>Class Purple</u>: This is the class of customers that earns low income but makes high purchases. They are the loyal customers who patronises the organisation depsite their level of income. They should receive some incentives and appreciation for their loyalty.
- <u>Class Green</u>: This is the class of customers that are average income earners and also makes average purchases. 
- <u>Class Blue</u>: This is the high income earning class of customers who makes low purchases. Advertisement should be targetted at this customers to enale buy more product as they have the financial backup to fund the purchases of the advertise product.

They are loyal to the organisation , as such, promotions and advertisement should be targetted at them, as well rewards and incentives for patrony so as to encourage them to increase their patrony.

```python
# Add the labels on the dataframe 
cluster_labels = kmeans_5.fit_predict(x)
frame['K_Means'] = cluster_labels
frame['K_Means'] = frame.K_Means.replace({4: 'lw_inc and lw_pur', 2:'md_inc and md_pur', 3:'hi_inc and hi_pur', 1:'hi_inc and lw_pur', 5:'lw_inc and hi_pur'})
```

<h1><b><u>HIERARCHICHAL CLUSTERING</u></b></h2>

```python
# We endeavour to identify the numbers of Hierarchy needed for the clustering

import scipy.cluster.hierarchy as shc  
plt.figure(figsize = (18, 7))

dendro = shc.dendrogram(shc.linkage(x, method="ward"))  
plt.title("Dendogram Plot")  
plt.ylabel("Euclidean Distances")  
plt.xlabel("Customers")  
plt.show()  
```

Using this Dendrogram, we will now determine the optimal number of clusters for our model. For this, we will find the maximum vertical distance that does not cut any horizontal bar. The maximum vertical line without any horizontal cut is the first line.... Cutting this line, we will have 5 clusters.

```python
# Train the Hierarchichal clustering on our dataset, having gotten the number of clusters

from sklearn.cluster import AgglomerativeClustering  
h_cluster = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')  
pred = h_cluster.fit_predict(x) 
```

```python
#visulaizing the clusters 
plt.figure(figsize = (12, 6)) 
plt.scatter(X[pred == 0, 0], X[pred == 0, 1], s = 100, c = 'blue', label = 'Cluster 1')  
plt.scatter(X[pred == 1, 0], X[pred == 1, 1], s = 100, c = 'green', label = 'Cluster 2')  
plt.scatter(X[pred== 2, 0], X[pred == 2, 1], s = 100, c = 'red', label = 'Cluster 3')  
plt.scatter(X[pred == 3, 0], X[pred == 3, 1], s = 100, c = 'cyan', label = 'Cluster 4')  
plt.scatter(X[pred == 4, 0], X[pred == 4, 1], s = 100, c = 'magenta', label = 'Cluster 5')  
plt.title('Clusters of customers')  
plt.xlabel('Income')  
plt.ylabel('Spending Score (1-100)')  
plt.legend()  
plt.show()  
```

```python
frame['H_cluster'] = pred
frame['H_cluster'] = frame.H_cluster.replace({4: 'lw_inc and hi_pur', 2:'md_inc and md_pur', 3:'hi_inc and hi_pur', 1:'hi_inc and lw_pur', 5:'lw_inc and lw_pur'})
frame.head()
```
