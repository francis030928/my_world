<h3><b><u>HOW DOES THE NEURAL NETWORK WORKS</u></b></h3>
Neural networks are a type of machine learning algorithm that can learn to recognize patterns in data. They're based on the way that the brain processes information, using simple computational units called neurons that are connected to each other in complex networks.
<br>
At the heart of a neural network are the weights and biases, which determine how strongly the inputs to each neuron are weighted and how easily that neuron will fire. When a neural network is initialized, the weights and biases are set to random values. During training, the network learns to adjust these values in order to improve its performance on the task it's being trained on.
<br>
To optimize the weights and biases, the network uses a technique called backpropagation. Backpropagation involves calculating the gradient of the loss function with respect to the weights and biases, and then using that gradient to update the weights and biases in the direction that will decrease the loss.
<br>
The loss function is a measure of how well the network is performing on the task it's being trained on. It's typically a function of the difference between the network's output and the true output for a given input.
<br>
Once the loss function has been calculated, the network backpropagates the gradient through the network, adjusting the weights and biases of each neuron in turn. This process continues until the weights and biases have been optimized to the point where the network is performing well on the task it's being trained on.
<br>
So in summary, neural networks work by initializing random weights and biases, optimizing them using backpropagation and a loss function, and updating them based on the gradient of the loss function to improve the network's performance.
Optimizers are algorithms used in neural networks to update the weights and biases of the neural network during the training process. They are used to minimize the loss function and find the optimal set of weights and biases that lead to the best performance of the neural network.
<br>
These are the basic steps the Neural Network takes to make generalizations:
<ol>
<li>Initialization: The weights and biases of the neural network are initialized with random values.</li>
<li>Forward Propagation: The input data is passed through the network, the weighted sum of inputs is calculated in each node, and the output of the activation function is computed.</li>
<li>Error Calculation: The output of the network is compared with the actual output to compute the error or loss.</li>
<li>Backpropagation: The error is propagated backwards through the network to update the weights and biases, which is done using an optimization algorithm like stochastic gradient descent (SGD). The weights are updated based on the partial derivatives of the loss with respect to the weights.</li>
<li>Repeat: The above steps are repeated for multiple epochs or iterations until the network achieves a satisfactory level of performance.
During the backpropagation step, the weights are updated in the direction that reduces the error or loss function. The learning rate and momentum are hyperparameters that affect the magnitude and direction of the weight updates.</li>
</ol>
The optimization process involves finding the values of the weights and biases that minimize the loss function. The choice of loss function depends on the problem being solved. For example, mean squared error (MSE) loss is commonly used for regression problems, while binary cross-entropy loss is used for binary classification problems.

<h3><u><b>What activation should we use, and When</b></u></h3>
<ol>
<li>For binary classification problems where the output is a probability value between 0 and 1, the most commonly used activation function is the sigmoid function. The sigmoid function maps any input value to a probability value between 0 and 1, which is ideal for binary classification problems where the goal is to predict the probability of belonging to a particular class.</li>
<li>For multiclass classification problems, where the output is a probability distribution over multiple classes, the most commonly used activation function for the final output layer is the softmax function. The softmax function is used to normalize the output vector so that it represents a probability distribution over the classes. The predicted class is then the one with the highest probability. </li>
<li>For regression problems, the most commonly used activation function is the linear activation function, which simply outputs the input value as is. This is because the goal of regression is to predict continuous numerical values, and the linear activation function allows the model to output values within a continuous range. </li>
<li>For intermediate layers, other activation functions such as ReLU (Rectified Linear Unit), LeakyReLU, and Tanh are commonly used in deep learning models. These functions help to introduce non-linearities to the model and improve its ability to learn complex relationships between the input features and the output.</li>
</ol>
Learning Rate: How much should this step outcome affect our eight and bias?<br>
Momentum: How should past outcomes affect our weight and bias?<br>

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import tensorflow as tf
from sklearn.preprocessing import LabelEncoder
```

```python
from sklearn.model_selection import train_test_split

# Load the Fashion MNIST dataset
fashion_mnist = tf.keras.datasets.fashion_mnist
(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()

# Split the dataset into training and validation sets
train_images, train_validation_images, train_labels, train_validation_labels = train_test_split(train_images, train_labels, test_size=0.16666, random_state=42)

# Print the shapes of the resulting datasets
print("Training set:", train_images.shape, train_labels.shape)
print("Validation set:", train_validation_images.shape, train_validation_labels.shape)
print("Test set:", test_images.shape, test_labels.shape)



# OR
# image_dataframe = tf.keras.datasets.fashion_mnist
# (train_images, train_labels), (test_images, test_labels) = image_dataframe.load_data()

# # Construct a train and validation set from the training set.

# validation_images = train_images[:10000, :, :] #................................ Selected (start from 0, and end at 10,000)
# validation_labels = train_labels[:10000] 

# # Construct a training set from the first 50000 images and labels.
# train_images = train_images[10000:, :, :] #..................................... Selected (start from 10,000 and end at the last digit)
# train_labels = train_labels[10000]

# # The images are 28x28 NumPy arrays, with pixel values ranging from 0 to 255. The labels are an array of integers, ranging from 0 to 9.
```

```python
# The classname of the Dataset
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']

# Each image is mapped to a single label inside this list above.
```

<h2><b><u>DATA EXPLORATION</u></b></h2>

```python
# Show the train data at index position 0
plt.figure()
index_rand = np.random.randint(0, len(train_images))
plt.imshow(train_images[index_rand])
numerical_label = train_labels[index_rand]
text_description = class_names[numerical_label]
plt.title('True Class: {} ("{}")'.format(numerical_label, text_description))
plt.colorbar()
plt.show()
```

<h2><b><u>DATA PREPROCESSING</u></b></h2>

```python
# Standard Scale the train and test image set
# Scale these values to a range of 0 to 1 before feeding them to the neural network model. 
# There are 255 pixels, hence to scale, divide the values by 255. 

train_images = train_images / 255.0
test_images = test_images / 255.0
```

```python
# Display the classes in the train image sets

plt.figure(figsize=(18,10))
sns.set(style = 'darkgrid')
for i in range(35):
    plt.subplot(5,7,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    rand_int = np.random.randint(0, 50000)
    plt.imshow(train_images[rand_int])
    plt.xlabel(class_names[train_labels[rand_int]])
plt.show()
```

<h2><b><u>MODELLING</u></b></h2>

```python
model = tf.keras.Sequential([ #........................ Instantiate the model creating class.
    tf.keras.layers.Flatten(input_shape=(28, 28)), #... Flatten out the image from 28 by 28 pixels to a one dimensional 784 pixels
    tf.keras.layers.Dense(128, activation='relu'), #... Add the first 128 layer, and instantiate the activation to be used.
    tf.keras.layers.Dense(10, activation='sigmoid')#.The second (last) layer which comprises of the number of possible outcomes. 10 outcomes in this case
])
model.compile(optimizer='adam', # ..................... The optimizer that adjusts weight and bias for a given neuron
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True), #...... Loss calculates the error of the prediction
              metrics=['accuracy']) #.................. Accuracy calculates the precision of the prediction.

model.fit(train_images, train_labels, epochs=10) #..... Fit the model on the dataset and define the number of epochs
```

```python
# Visualize the created model

from tensorflow.keras.utils import plot_model
plot_model(model,
           to_file="model.png",
           show_shapes=True,
           show_layer_names=True,
          )
```

```python
# Evaluate the performance
test_loss, test_acc = model.evaluate(test_images,  test_labels, verbose=2)
print('\nTest accuracy:', test_acc)
```

```python
# Make predictions for each image in the test image set.
probability_model = tf.keras.Sequential([model, 
                                         tf.keras.layers.Softmax()])
predictions = probability_model.predict(test_images)
print(predictions[0]) #.... This will print out 10 arrays of prediction which corresponds to the 10 classes of images
maximum = np.argmax(predictions[0])

print(f"\n The model predicted that first test image is of class {np.argmax(predictions[0])}")
```

```python
data = pd.read_csv('/content/BankChurners.csv')
pd.set_option('display.max_columns', None)
data.drop(['CLIENTNUM', 'Income_Category',
           'Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_1',
       'Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_2'],
       axis = 1, inplace = True)
data.head(3)
```

```python
data.dropna(inplace = True)
print(f'dataframe rows: {data.shape[0]} \ndataframe columns: {data.shape[1]} \n')

cats = data.select_dtypes(include = ['object', 'category'])
nums = data.select_dtypes(include = 'number')

display(cats.head(3))
display(nums.head(3))
```

```python
# Transform the dataset to a Machine-Readeable Language

frame = data.copy()

from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
for i in cats:
  frame[i] = encoder.fit_transform(frame[i])

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
for i in nums:
  frame[[i]] = scaler.fit_transform(frame[[i]])

print(f"Transformed Dataset")
display(frame.head())

y = frame.Attrition_Flag
x = frame.drop(['Attrition_Flag'], axis = 1)
```

```python
# - Using XGBOOST to find feature importance
import xgboost as xgb
model = xgb.XGBClassifier()
model.fit(x, y)

# Print feature importance scores
xgb.plot_importance(model)
```

```python
# ANOVA F-value between label/feature
from sklearn.feature_selection import SelectKBest, chi2, f_classif

best_feature1 = SelectKBest(score_func = f_classif, k = 15)
fitting1 = best_feature1.fit(x,y)
scores1 = pd.DataFrame(fitting1.scores_)
columns1 = pd.DataFrame(x.columns)
feat_score1 = pd.concat([columns1, scores1], axis = 1)
feat_score1.columns = ['Feature', 'F_classif_score'] 
k1 = feat_score1.nlargest(20, 'F_classif_score')

k1.sort_values(by = 'F_classif_score', ascending = False)
```

```python
sel_features = ['Total_Trans_Ct', 'Total_Ct_Chng_Q4_Q1', 'Total_Revolving_Bal',
       'Contacts_Count_12_mon', 'Avg_Utilization_Ratio',
       'Total_Trans_Amt', 'Months_Inactive_12_mon',
       'Total_Relationship_Count', 'Total_Amt_Chng_Q4_Q1', 'Gender',
       'Credit_Limit', 'Dependent_count']

new_frame = x[sel_features] #................................................... Turn the selected features to a dataframe
new_frame.head() #.............................................................. Display the selected features
```

```python
# - Check for Multi-Colinearity in the chosen features.
plt.figure(figsize = (14, 7))
sns.heatmap(new_frame.corr(), annot = True, cmap = 'GnBu')
```

<h2><b><u>MODELLING</u></b></h2>

```python
# - split into train and test
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report


x_train, x_test, y_train, y_test = train_test_split(new_frame, y, test_size = 0.15, random_state = 47, stratify = y)
print(f'x_train: {x_train.shape}')
print(f'x_test: {x_test.shape}')
print('y_train: {}'.format(y_train.shape))
print('y_test: {}'.format(y_test.shape))
```

```python
model = tf.keras.Sequential([ #........................ Instantiate the model creating class.
    tf.keras.layers.Dense(units=12, activation='relu'), #... Input layer of 12 features
    tf.keras.layers.Dense(20, activation='relu'), #.... Add the second 20 layer, and instantiate the activation to be used.
    tf.keras.layers.Dense(15, activation='relu'), #..... Add the third layer.
    tf.keras.layers.Dense(1, activation='sigmoid') #... Add the last output layer
])
model.compile(optimizer='adam', # ..................... The optimizer that adjusts weight and bias for a given neuron
              loss = 'binary_crossentropy', #...... Loss calculates the error of the prediction
              metrics=['accuracy']) #.................. Accuracy calculates the precision of the prediction.

model.fit(x_train, y_train, epochs=10) #..... Fit the model on the dataset and define the number of epochs
```

```python
from tensorflow.keras.utils import plot_model
plot_model(model,
           to_file="model.png",
           show_shapes=True,
           show_layer_names=True,
          )


#Saving created neural network
model.save("neural_network_model.h5")
```

```python
y_pred = model.predict(x_test)
y_pred = (y_pred > 0.5)  #............................................... set a 50% confidence level that the customer doesnt stop buying
outcome = pd.DataFrame(list(zip(y_test, y_pred)), columns=['Actual', 'Predicted'])
outcome


# The customers having TRUE are existing and keep buying, while FALSE are attrited and will stop buying
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.metrics import classification_report

revealer = confusion_matrix(y_pred, y_test)
sns.set(style = 'darkgrid')
sns.heatmap(revealer/np.sum(revealer), annot=True, cmap='crest', fmt='.1%', linewidth=1)

print(classification_report(y_pred, y_test))
```
